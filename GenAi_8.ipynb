```python
import os
from google.colab import userdata
from TTS.api import TTS

# Load Hugging Face token from Colab secrets
HF_TOKEN = userdata.get("HF_TOKEN")

if HF_TOKEN is None:
    raise ValueError("Hugging Face token not found. Make sure you've added 'HF_TOKEN' to Colab secrets.")

# Set token for Hugging Face access
os.environ["HUGGINGFACEHUB_API_TOKEN"] = HF_TOKEN
```

```python
# Load pre-trained TTS model
tts = TTS("tts_models/en/ljspeech/tacotron2-DDC")

# Input text
text = "Hello, how are you today?"

# Generate speech and save to file
tts.tts_to_file(text=text, file_path="output.wav")

print("✅ Speech generated and saved as output.wav")
```
